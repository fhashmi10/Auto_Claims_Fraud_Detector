```python
import pandas as pd
df = pd.read_csv('../data/insurance_claims.csv')
df.head(2)
```

```python
# drop cols - incident location (as already have city and state)
df.drop('incident_location', axis=1, inplace = True)
df.head(1)

# drop empty col
df.drop('_c39', axis=1, inplace = True)
```

```python
# separate out policy csl
df[['policy_csl_low', 'policy_csl_high']] = df['policy_csl'].str.split('/', expand=True)
df[["policy_csl_low", "policy_csl_high"]] = df[["policy_csl_low", "policy_csl_high"]].apply(pd.to_numeric)
df.drop('policy_csl', axis=1, inplace = True)

df.info()
```

```python
# handle datetime
# difference between these two dates policy_bind_date and incident_date
df[["policy_bind_date", "incident_date"]] = df[["policy_bind_date", "incident_date"]].apply(pd.to_datetime)
df.info()
```

```python
df["incident_days_since_inception"]=(df["incident_date"] - df["policy_bind_date"])/pd.Timedelta(days=1)
df["incident_days_since_inception"].apply(pd.to_numeric)
df.drop('incident_date', axis=1, inplace = True)
df.drop('policy_bind_date', axis=1, inplace = True)
df.info()
```
